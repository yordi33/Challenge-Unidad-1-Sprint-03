## 1. ¿Qué es Git?

Git es un **sistema de control de versiones distribuido** creado por Linus Torvalds en 2005. Permite:

- Registrar el historial completo de cambios de un proyecto
- Trabajar en paralelo sin pisarse con otros colaboradores
- Volver a cualquier punto anterior del proyecto
- Colaborar con miles de personas en el mismo código

> **Git ≠ GitHub.** Git es la herramienta que corre en tu máquina. GitHub, GitLab y Bitbucket son plataformas online que alojan repositorios Git y añaden funcionalidades de colaboración.


### Repositorio Git

Un repositorio (o "repo") es la carpeta que Git vigila. Contiene:
- Todos los archivos del proyecto
- La carpeta oculta `.git/` con toda la historia

Cada repositorio es independiente: puede vivir solo en tu máquina, o sincronizarse con uno remoto (GitHub, etc.).

## 2. Flujo de trabajo fundamental

El flujo básico de Git tiene tres zonas:

```
Directorio de trabajo  →  Staging Area  →  Repositorio (.git)
   (tus archivos)           git add           git commit
```

### Staging Area

La **staging area** (o índice) es un espacio intermedio donde preparas los cambios antes de confirmarlos. Esto te da control fino: puedes tener 10 archivos modificados y hacer commit solo de 3.

```bash
git add archivo.txt          # añade un archivo concreto
git add .                    # añade todos los cambios del directorio
git add -p                   # modo interactivo: elige qué partes añadir
```

### Commit

Un **commit** es una instantánea permanente del proyecto en ese momento. Cada commit tiene:
- Un **hash SHA-1** único (ej: `85a84ef7b8b261ea680308d7f2d672f161e8a01a`)
- Un **mensaje** que describe qué cambia y por qué
- El autor, la fecha, y un puntero al commit anterior

```bash
git commit -m "feat: add user login form"
git commit                   # abre el editor para un mensaje largo
git commit --amend           # modifica el último commit (¡solo si no has hecho push!)
```

### HEAD

**HEAD** es un puntero que indica dónde estás ahora mismo en la historia del repositorio. Normalmente apunta a la rama activa.

```
commit C ← commit B ← commit A
                           ↑
                          HEAD (main)
```

## 3. Primeros comandos

### Configuración inicial (solo una vez)

```bash
git config --global user.name "Tu Nombre"
git config --global user.email "tu@email.com"
git config --global core.editor "code --wait"   # VS Code como editor
git config --list                                # ver toda la configuración
```

### Crear e inspeccionar un repositorio

```bash
git init                    # inicia un repo en la carpeta actual
git status                  # estado actual: modified, staged, untracked
git add .                   # añade todo al staging
git commit -m "mensaje"     # guarda un commit
```

### .gitignore

El archivo `.gitignore` indica a Git qué archivos **nunca** debe rastrear. Es imprescindible en cualquier proyecto real.

```
# .gitignore
node_modules/
.env
*.log
dist/
.DS_Store
__pycache__/
```

> Crea tu `.gitignore` desde el inicio del proyecto. [gitignore.io](https://gitignore.io) genera uno automáticamente según tu stack.


###  Ejercicio 1: Tu primer repositorio

1. Crea una carpeta llamada `mi-proyecto`.
2. Entra en ella y ejecuta `git init`.
3. Comprueba con `git status` que no estás dentro de otro repo.
4. Crea un archivo `.gitignore` con `*.log` dentro.
5. Crea `README.md` con una línea de texto.
6. Añade ambos al staging con `git add .`.
7. Haz el primer commit: `git commit -m "chore: initial project setup"`.
8. Verifica el resultado con `git log`.

## 4. Historial y diferencias

### Ver el historial

```bash
git log                          # historial completo
git log --oneline                # una línea por commit
git log --oneline --graph --all  # árbol visual de todas las ramas 
git log --author="Nombre"        # filtrar por autor
git log --since="2024-01-01"     # filtrar por fecha
```

### Ver diferencias

```bash
git diff                         # cambios en el directorio de trabajo (sin stage)
git diff --staged                # cambios en el staging (listos para commit)
git diff main..develop           # diferencia entre dos ramas
git show abc1234                 # ver un commit concreto por su hash
```

### Deshacer cambios

```bash
git restore archivo.txt          # descarta cambios en el directorio ( irreversible)
git restore --staged archivo.txt # saca un archivo del staging (sin borrar cambios)
git revert abc1234               # crea un nuevo commit que deshace otro (seguro en equipo)
git reset --soft HEAD~1          # deshace el último commit, mantiene cambios en staging
```

>  `git reset --hard` borra cambios permanentemente. Úsalo solo si sabes lo que haces.

## 5. Ramas (Branches)

Las ramas son **líneas de desarrollo paralelas**. Permiten trabajar en una nueva funcionalidad sin tocar el código estable.

```
main:    A ── B ── C ──────────── F
                    \            /
feature:            C ── D ── E
```

### Comandos de ramas

```bash
git branch                       # lista ramas locales
git branch -a                    # lista todas (locales + remotas)
git branch nueva-rama            # crea una rama (no cambia a ella)

# Cambiar de rama
git switch nombre-rama           # cambia a una rama existente
git switch -c nueva-rama         # crea Y cambia de una vez 

# Alternativa cambio de ramas (sigue funcionando)
git checkout nombre-rama
git checkout -b nueva-rama

# Eliminar ramas
git branch -d rama               # elimina (solo si ya fue fusionada)
git branch -D rama               # fuerza la eliminación
```



### Nombrar ramas

Una convención habitual para nombrar ramas:

| Tipo | Ejemplo |
|------|---------|
| Nueva funcionalidad | `feature/user-login` |
| Corrección de bug | `fix/navbar-overflow` |
| Corrección urgente en producción | `hotfix/payment-crash` |
| Preparación de versión | `release/v2.1.0` |
| Experimento | `experiment/new-animation` |


###  Ejercicio 2: Ramas Patronus

1. Crea una carpeta `Patronus` e inicia un repositorio Git.
2. Crea `patronus.txt` vacío y haz commit: `"chore: add empty patronus file"`.
3. Crea dos ramas desde `main`: `harry` y `snape`.
4. Cambia a `harry` con `git switch harry`.
5. Añade el arte ASCII del ciervo de Harry en `patronus.txt`.
6. Haz commit: `"feat: add Harry's stag patronus"`.

8. Añade el arte ASCII del ciervo de Snape.

11. Cambia el título a "PATRONUS DE LILY".

13. Ejecuta `git log --oneline --graph --all` para ver el árbol de ramas.
14. **Bonus**: elimina la rama `snape` con `git branch -d snape`.

## 6. Fusión de ramas (Merge)

Fusionar integra los cambios de una rama en otra. Siempre **te fusionas hacia la rama activa (HEAD)**.

```bash
git switch main              # primero ve a la rama destino
git merge feature/login      # trae los cambios de feature/login a main
```

### Tipos de merge

**Fast-forward**: cuando la rama destino no tiene commits nuevos desde la bifurcación. Git simplemente mueve el puntero.

```
Antes:   main: A ── B
                     \
         feature:    B ── C ── D

Después: main: A ── B ── C ── D   ← solo movió el puntero
```

**Merge commit**: cuando ambas ramas tienen commits distintos. Git crea un commit nuevo que une ambas historias.

```
Antes:   main:    A ── B ── E
                   \
         feature:  B ── C ── D

Después: main:    A ── B ── E ── M   ← M es el merge commit
                   \          /
         feature:  B ── C ── D
```

```bash
git merge --no-ff feature/login  # fuerza merge commit aunque sea posible fast-forward
git merge --squash feature/login # aplasta todos los commits en uno antes de fusionar
```


### Ejercicio 3: Fusionando el Patronus

3. Fusiona la rama `lily`: `git merge lily`.

## 7. Resolución de conflictos

Un **conflicto** ocurre cuando dos ramas modifican las mismas líneas de un archivo. Git no puede decidir cuál versión prevalece: te pide que lo hagas tú.

### ¿Cómo se ve un conflicto?

```
<<<<<<< HEAD
Esta es la versión de la rama actual (main)
=======
Esta es la versión de la rama que estás fusionando (feature/x)
>>>>>>> feature/x
```

### Pasos para resolver un conflicto

```bash
# 1. Intenta el merge
git merge feature/x

# 2. Git te avisa de los conflictos
# CONFLICT (content): Merge conflict in archivo.txt

# 3. Abre el archivo y edítalo manualmente (quédate con lo correcto)
#    Elimina los marcadores <<<<<<, =======, >>>>>>>

# 4. Marca el conflicto como resuelto
git add archivo.txt

# 5. Completa el merge
git commit
```

> 💡 La mayoría de editores (VS Code, IntelliJ...) tienen interfaz visual para resolver conflictos. Busca "merge editor" en tu editor.

### Git stash: guardar trabajo sin hacer commit

Cuando necesitas cambiar de rama pero tienes trabajo a medias:

```bash
git stash                    # guarda los cambios temporalmente
git switch otra-rama         # trabaja en otra cosa
git switch mi-rama           # vuelve
git stash pop                # recupera los cambios guardados
git stash list               # lista todos los stash guardados
```

## 8. Trabajo remoto

Un **repositorio remoto** es una copia del repo en otro lugar (GitHub, GitLab, etc.). Permite colaborar en equipo y tener una copia de seguridad.

### Conectar con un remoto

```bash
git remote add origin https://github.com/usuario/repo.git
git remote -v                           # ver remotos configurados
git remote remove origin                # eliminar un remoto
```

### Clonar un repositorio existente

```bash
git clone https://github.com/usuario/repo.git
git clone https://github.com/usuario/repo.git mi-carpeta   # con nombre propio
```

### Subir y bajar cambios

```bash
git push origin main                    # sube la rama main al remoto
git push -u origin main                 # sube y configura seguimiento (primera vez)
git push                                # si ya hay seguimiento configurado

git pull                                # fetch + merge en un paso
git pull origin main                    # equivalente explícito

git fetch origin                        # descarga cambios SIN aplicarlos
git fetch origin main                   # solo una rama concreta
```

### Fetch vs Pull

| Comando | ¿Qué hace? | ¿Cuándo usarlo? |
|---------|-----------|-----------------|
| `git fetch` | Descarga cambios del remoto sin tocar tu directorio de trabajo | Cuando quieres ver qué hay nuevo antes de integrarlo |
| `git pull` | Fetch + merge automático | Cuando quieres actualizar tu rama directamente |

>  En equipos es buena práctica hacer `git fetch` primero, revisar los cambios con `git log origin/main`, y luego hacer el merge manualmente.

### Ramas remotas

```bash
git branch -r                           # ver ramas remotas
git branch -a                           # locales + remotas
git switch -c feature/x origin/feature/x   # crear rama local desde una remota
git push origin --delete rama           # eliminar rama remota
```

## 9. Git & GitHub: colaboración real

### Git vs GitHub

| | Git | GitHub |
|--|-----|--------|
| Qué es | Herramienta de control de versiones | Plataforma de colaboración |
| Dónde corre | En tu máquina (local) | En la nube (online) |
| Alternativas | — | GitLab, Bitbucket, Gitea |

### Pull Request (PR)

Un **Pull Request** es una propuesta de cambio: pides que tus cambios en una rama se fusionen en otra. Es el mecanismo central de la colaboración en GitHub.

Flujo típico:
1. Haces fork del repo (si no tienes acceso directo) o creas una rama
2. Trabajas en tu rama y subes los commits
3. Abres un PR en GitHub
4. Tus compañeros revisan el código y comentan
5. Corriges si hace falta y el PR se aprueba
6. Se hace merge a la rama principal

### Funcionalidades clave de GitHub

- **Issues**: seguimiento de tareas, bugs y mejoras
- **GitHub Actions**: automatización de tests, deploy, etc. (CI/CD)
- **GitHub Pages**: hosting gratuito de sitios estáticos
- **Discussions**: foro de conversación del proyecto
- **Projects**: tableros kanban para organizar el trabajo
- **Gist**: compartir fragmentos de código sueltos

### Fork

Un **fork** es una copia completa de un repositorio en tu cuenta de GitHub. Permite contribuir a proyectos en los que no tienes permisos de escritura directa.

```bash
# Flujo con fork
# 1. Fork en GitHub (botón en la web)
# 2. Clona tu fork
git clone https://github.com/TU_USUARIO/repo-forkeado.git

# 3. Añade el repo original como remoto "upstream"
git remote add upstream https://github.com/AUTOR_ORIGINAL/repo.git

# 4. Trabaja en una rama, haz push a tu fork
git switch -c mi-contribucion
git push origin mi-contribucion

# 5. Abre PR desde tu fork al repo original
```

## 10. Flujo de trabajo profesional

En equipos reales se suelen seguir variantes de **Git Flow** o **GitHub Flow**. Aquí el modelo más habitual:

### Ramas permanentes

| Rama | Propósito |
|------|-----------|
| `main` | Código en producción. Solo recibe merges via PR. |
| `develop` | Integración de features. Se despliega a staging. |

### Ramas temporales

| Tipo | Formato | Desde | Fusionar a |
|------|---------|-------|------------|
| Nueva funcionalidad | `feature/nombre` | `develop` | `develop` |
| Corrección de bug | `fix/nombre` | `develop` | `develop` |
| Hotfix en producción | `hotfix/nombre` | `main` | `main` + `develop` |
| Preparar lanzamiento | `release/vX.Y.Z` | `develop` | `main` + `develop` |

### Flujo típico de una feature

```bash
# 1. Actualizar develop
git switch develop
git pull origin develop

# 2. Crear rama de feature
git switch -c feature/nueva-funcionalidad

# 3. Desarrollar con commits atómicos
git add .
git commit -m "feat(módulo): descripción clara"

# 4. Subir y abrir PR
git push origin feature/nueva-funcionalidad
# → Abrir Pull Request en GitHub: feature/nueva-funcionalidad → develop

# 5. Tras el merge, limpiar
git switch develop
git pull origin develop
git branch -d feature/nueva-funcionalidad
```

### Tags: marcar versiones

```bash
git tag v1.0.0                          # tag ligero
git tag -a v1.0.0 -m "Primera versión" # tag anotado (recomendado)
git push origin v1.0.0                  # subir el tag
git push origin --tags                  # subir todos los tags
```

## 11. Buenas prácticas

### Conventional Commits

El estándar [Conventional Commits](https://www.conventionalcommits.org) da formato consistente a los mensajes y permite generar CHANGELOGs automáticamente.

```
<tipo>(<ámbito opcional>): <descripción corta>

[cuerpo opcional]

[nota al pie opcional]
```

#### Tipos principales

| Tipo | Cuándo usarlo |
|------|---------------|
| `feat` | Nueva funcionalidad |
| `fix` | Corrección de bug |
| `chore` | Tareas de build, dependencias, configuración |
| `docs` | Solo documentación |
| `test` | Añadir o modificar tests |
| `refactor` | Refactorización sin cambio de comportamiento |
| `style` | Formato, espacios, punto y coma (sin lógica) |
| `perf` | Mejoras de rendimiento |
| `ci` | Cambios en CI/CD |

#### Ejemplos reales

```bash
git commit -m "feat(auth): add OAuth2 login with Google"
git commit -m "fix(cart): prevent negative item quantity"
git commit -m "chore: upgrade dependencies to latest"
git commit -m "docs: update API reference for v2 endpoints"
git commit -m "test(payments): add edge cases for failed charges"
git commit -m "feat!: redesign user profile API"  # ! = breaking change
```

### Commits atómicos

Cada commit debe representar **un solo cambio lógico**. Un buen commit:
- Es pequeño y enfocado
- El código funciona antes y después del commit
- El mensaje explica el *qué* y el *por qué*, no el *cómo*

### Checklist antes de hacer push

- [ ] `git status` limpio (nada inesperado)
- [ ] `git diff --staged` revisado
- [ ] Tests pasando
- [ ] Mensaje de commit claro y en formato Conventional Commits
- [ ] `.gitignore` actualizado si añadiste nuevos archivos ignorables

## 12. Ejercicios integradores

###  Ejercicio 5: Flujo de trabajo completo en equipo

Simula un flujo profesional en local (o con un compañero en GitHub):

1. Crea un repo `tienda-online` e inicialízalo con un `README.md`.
2. Crea la rama `develop` desde `main`.
3. Desde `develop`, crea `feature/productos`.
4. Añade un archivo `productos.txt` con 3 productos. Haz 2 commits:
   - `"feat(productos): add initial product list"`
   - `"feat(productos): add product descriptions"`
5. Fusiona `feature/productos` en `develop` (simulando un PR).
6. Desde `develop`, crea `feature/carrito`.
7. Añade `carrito.txt`. Haz commit: `"feat(carrito): add shopping cart logic"`.
8. **Simula un hotfix**: desde `main`, crea `hotfix/precio-erroneo`.
   - Modifica `productos.txt`.
   - Commit: `"fix(productos): correct price for product A"`.
   - Fusiona el hotfix en `main` (y también en `develop`).
9. Etiqueta la versión en main: `git tag -a v1.0.1 -m "hotfix: correct price"`.
10. Fusiona `feature/carrito` en `develop`.
11. Fusiona `develop` en `main`. Etiqueta: `git tag -a v1.1.0 -m "release: add cart feature"`.
12. Visualiza toda la historia: `git log --oneline --graph --all`.

###  Referencia rápida de comandos

```bash
# CONFIGURACIÓN
git config --global user.name "Nombre"
git config --global user.email "email"

# REPOSITORIO
git init                          # iniciar repo
git clone <url>                   # clonar repo remoto
git status                        # ver estado actual

# STAGING Y COMMITS
git add <archivo>                 # añadir al staging
git add .                         # añadir todo
git commit -m "mensaje"           # crear commit
git commit --amend                # modificar último commit

# HISTORIAL
git log --oneline --graph --all   # árbol de commits
git diff                          # cambios sin stage
git diff --staged                 # cambios en stage

# RAMAS
git branch                        # listar ramas
git switch -c <rama>              # crear y cambiar de rama
git switch <rama>                 # cambiar de rama
git merge <rama>                  # fusionar rama
git branch -d <rama>              # eliminar rama

# DESHACER
git restore <archivo>             # descartar cambios
git restore --staged <archivo>    # sacar del staging
git revert <hash>                 # deshacer commit (seguro)
git stash / git stash pop         # guardar/recuperar trabajo temporal

# REMOTO
git remote add origin <url>       # conectar remoto
git push -u origin main           # subir rama (primera vez)
git pull                          # bajar e integrar cambios
git fetch                         # bajar cambios sin integrar

# TAGS
git tag -a v1.0.0 -m "mensaje"   # crear tag anotado
git push origin --tags            # subir todos los tags
```
